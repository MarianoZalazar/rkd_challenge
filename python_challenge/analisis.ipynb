Como esta distribuida la cantidad de atletas hombre y mujeres a lo largo de las disciplinas?

En base a los porcentajes de ambos generos conseguimos una diferencia la cual utilizamos para establecer una etiqueta:
- M = Disciplina dominada por hombres
- F = Disciplina dominada por mujeres
- E = Disciplina equitativa

```python
ratios = gender
```

```python
ratios['DiffBetweenMaleFemale'] = ratios.Male - ratios.Female
```

```python
ratios['GenderDominance'] = np.where(ratios['DiffBetweenMaleFemale'] > 0, 'Male', np.where(ratios['DiffBetweenMaleFemale'] < 0, 'Female', 'Equal'))
```

```python
ratios['DiffBetweenMaleFemale'] = abs(ratios['DiffBetweenMaleFemale'])
```

```python
counts = ratios.groupby('GenderDominance', as_index=False).agg({'DiffBetweenMaleFemale':'mean', 'Discipline':'count'}).rename(columns={'Discipline':'DisciplineQty'})
```

El resultado de este analisis denota que la cantidad de disciplinas dominadas por hombres es mayor a las dominadas por mujeres asi como que las disciplinas equitativas son mayoria

```python
px.bar(counts, x='GenderDominance', y='DisciplineQty', color='GenderDominance')
```

Debajo podemos observar la media de la cantidad de participantes por la cual un genero domina un deporte

```python
px.bar(counts.loc[counts['GenderDominance']!='Equal'], x='GenderDominance', y='DiffBetweenMaleFemale', color='GenderDominance')
```

```python
ratios_noc = ratios.merge(noc_per_discipline, on='Discipline').rename(columns={'NOC':'Country Participants Per Discipline'})
```

Aca analizamos como estan distribuidos los deportes en base a la cantidad de paises que participan y la predominancia de alguno de los generos

```python
px.bar(ratios_noc, x='Discipline', y='Country Participants Per Discipline', color='GenderDominance')
```

#IDH (Indice de Desarrollo Humano)

```python
hdi_per_noc = (athletes.groupby('NOC', as_index=False).Name.count()
              .merge(hdi, how='left', left_on='NOC', right_on='Country')
              .drop(columns=['Country']).rename(columns={'Name':'AthletesQtyPerCountry'}))
```

```python
hdi_per_noc=hdi_per_noc.loc[hdi_per_noc['NOC']!="Refugee Olympic Team"]
hdi_per_noc.HDI.fillna(hdi_per_noc.HDI.mean(), inplace=True)
```

```python
hdi_disciplines_athletes = hdi_per_noc.merge(discipline_per_noc, how='left', on='NOC').rename(columns={'Discipline':'DisciplineQtyPerCountry'})
```

En este grafico podemos analizar como afecta el Indice de Desarrollo Humano de un pais a la cantidad de disciplinas en las que participa un pais en las Olimpiadas

```python
px.scatter(hdi_disciplines_athletes, x='DisciplineQtyPerCountry', y='HDI', hover_data=['NOC'])
```

```python
px.scatter(hdi_disciplines_athletes, x='AthletesQtyPerCountry', y='HDI', hover_data=['NOC'])
```

Podemos concluir en que el Indice de Desarrollo Humano afecta a la cantidad de oportunidades de un pais en las Olimpiadas

Tomamos como objeto de analisis la correlacion entre la cantidad de medallas por pais con distintas variables

```python
discipline_count = athletes.groupby('NOC', as_index=False).Discipline.nunique().sort_values('Discipline').rename(columns={'Discipline':'DisciplineQtyPerNOC'})
```

```python
athletes_counts = athletes.groupby('NOC', as_index=False).Name.count().sort_values('Name').rename(columns={'Name':'AthletesQtyPerNOC'})
```

```python
medals_discipline = medals.merge(discipline_count, left_on='Team/NOC', right_on='NOC').drop(columns='NOC')
```

```python
medals_discipline_corr = medals_discipline.corr()
```

Este grafico nos permite observar la correlacion entre la cantidad de medallas y los rankings con la cantidad de disciplinas a las que asistio un pais. (En el caso de Rank y Rank by Total, entre menor sea el valor mayor es el puesto)

```python
fig = go.Figure()
fig.layout.height = 600
fig.layout.width = 600
fig.add_trace(go.Heatmap(z=medals_discipline_corr.values, x=medals_discipline_corr.index.values, y=medals_discipline_corr.columns.values, colorscale=px.colors.diverging.RdBu))
```

```python
px.scatter(medals_discipline, x='Rank', y='DisciplineQtyPerNOC', hover_data=['Team/NOC'])
```

```python
medals_athletes = medals.merge(athletes_counts, left_on='Team/NOC', right_on='NOC').drop(columns='NOC')
```

```python
medals_athletes_corr = medals_athletes.corr()
```

En este caso analizamos como la cantidad de atletas enviados por un pais afecta a las medallas recibidas por el mismo. (En el caso de Rank y Rank by Total, entre menor sea el valor mayor es el puesto)

```python
fig = go.Figure()
fig.layout.height = 600
fig.layout.width = 600
fig.add_trace(go.Heatmap(z=medals_athletes_corr.values, x=medals_athletes_corr.index.values,y=medals_athletes_corr.columns.values,colorscale=px.colors.diverging.RdBu))
```

```python
px.scatter(medals_athletes, x='Rank', y='AthletesQtyPerNOC', hover_data=['Team/NOC'])
```

```python
medals_hdi = hdi_per_noc.merge(medals, left_on='NOC', right_on='Team/NOC').drop(columns='NOC')
```

```python
medals_hdi_corr = medals_hdi.corr()
```

Para finalizar con estas correlaciones analizamos como afecta el Indice de Desarrollo Humano de cada pais a las medallas ganadas. (En el caso de Rank y Rank by Total, entre menor sea el valor mayor es el puesto)

```python
fig = go.Figure()
fig.layout.height = 600
fig.layout.width = 600
fig.add_trace(go.Heatmap(z=medals_hdi_corr.values,x=medals_hdi_corr.index.values,y=medals_hdi_corr.columns.values,colorscale=px.colors.diverging.RdBu))
```

```python
medals_discipline['MedalsPerDiscipline'] = medals_discipline.Total / medals_discipline.DisciplineQtyPerNOC
```

```python
medals_athletes['MedalsPerAthletes'] = medals_athletes.Total / medals_athletes.AthletesQtyPerNOC
```

```python
medals_discipline.loc[medals_discipline.Total > 5].sort_values('MedalsPerDiscipline', ascending=False).reset_index()[['Rank', 'Team/NOC', 'Total', 'DisciplineQtyPerNOC', 'MedalsPerDiscipline']]
```

```python
medals_athletes.loc[medals_athletes.Total > 5].sort_values('MedalsPerAthletes', ascending=False).reset_index()[['Rank', 'Team/NOC', 'Total', 'AthletesQtyPerNOC', 'MedalsPerAthletes']]
```

En este grafico observamos a los paises que tienen mejor promedio de medallas por cantidad de disciplinas en las que participaron
